```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('d.csv')

# Extract the required columns
categories = data['school classification']
admitted = data['number of admitted']
max_score = data['maximum test score']
min_score = data['minimum test score']
avg_score = data['average score']

plt.figure(figsize=(10, 6))
plt.plot(categories, max_score, marker='o', label='Max Score')
plt.plot(categories, min_score, marker='o', label='Min Score')
plt.plot(categories, avg_score, marker='o', label='Avg Score')
plt.xlabel('School Classification')
plt.ylabel('Score')
plt.title('Ala-too International University admittion')
plt.legend()
plt.savefig('diagram1.png', dpi=300)

# Display the chart
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('literacy2.csv')

# Select the columns for plotting
columns_to_plot = ['15+ female', '15+ male', '15+ total', '15-24 female', '15-24 male', '15-24 total',
                   '25-64 female', '25-64 male', '25-64 total', '65+ female', '65+ male', '65+ total']

# Plotting
plt.figure(figsize=(12, 8))

# Set colors for lines
colors = ['blue', 'orange', 'green', 'red', 'purple', 'gray',
          'cyan', 'magenta', 'yellow', 'darkblue', 'darkorange', 'darkgreen']

# Plot each line
for i, col in enumerate(columns_to_plot):
    plt.plot(data['Country'], data[col], marker='o', color=colors[i], label=col)

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Literacy Rate')
plt.title('Comparison of Literacy Rates in Central Asian Countries')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('diagram2.png', dpi=300)

plt.show()
```
